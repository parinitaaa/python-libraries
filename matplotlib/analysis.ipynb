```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y)
plt.show()
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y,marker='s')
plt.show()
plt.plot(x,y,marker='s',markersize=30) #markersize or ms
plt.show()
plt.plot(x,y,marker='s',markersize=10,markerfacecolor='red') #markerfacesize or mfs
plt.show()
plt.plot(x,y,marker='s',markersize=10,markerfacecolor='red',markeredgecolor='yellow') #markeredgesize or mes
plt.show()
```
