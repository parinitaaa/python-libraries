```python
#filtering:pokemons over 2metres of height
print('pokemons over 2m of height')
tall_pokemon=df[df['Height']>=2]
print(tall_pokemon)
print()
print('pokemons over 100kg of weight')
heavy_pokemon=df[df['Weight']>=100] #over 100kgs of weight
print(heavy_pokemon)
print()
print('legendary pokemons are')
legendary_pokemon=df[df['Legendary']==1] #1 or True - same thing
print(legendary_pokemon)
print()
print('water pokemons are')
water_pokemon=df[(df['Type1']=='Water') | (df['Type2']=='Water' )] #python-'or' logical-'||' but here we're using c style that is '|'
print(water_pokemon)
print()
print("fire and flying pokemons are:")
ff=df[(df['Type1']=='Fire') & (df['Type2']=='Flying' )]
print(ff)
    
```

```python
#Aggregation: reducing set of values to single summary value; used to summarize and analyse data

#aggregate functions for whole data frame
print("mean:")
print(df.mean(numeric_only=True))
print()
print("sum:")
print(df.sum(numeric_only=True))
print()
print("min:")
print(df.min(numeric_only=True))
print()
print("max:")
print(df.max(numeric_only=True))
print()
print("count:")
print(df.count())
print()
```

```python
#aggregate functions for single column - height
print("height column")
print("mean:")
print(df['Height'].mean())
print()
print("sum:")
print(df['Height'].sum())
print()
print("min:")
print(df['Height'].min())
print()
print("max:")
print(df['Height'].max())
print()
print("count:")
print(df['Height'].count())
print()
 
```

```python
#groupby function

 group=df.groupby("Type1")
print(group['Height'].mean())
print()
print(group['Height'].count())
```

```python
#data cleaning:process of fixing/removing incomplete, incorrect or irrelevant data
```

```python
#1. removing unnecessary columns
df=df.drop(columns=['No']) #alr dropped legendary, now dropping number columns
df
 
```

```python
#2.dropping rows with unknown values
q=pd.read_csv('data.csv')
q=q.dropna(subset=['Type2'])
q
```

```python
#3.replacing missing values
qr=pd.read_csv('data.csv')
qr=qr.fillna({'Type2':"None"})
qr
```

```python
#4.inconsistent values
p=pd.read_csv('data.csv')
p['Type1']=p['Type1'].replace({"Grass":"GRASS","Fire":"FIRE","Dragon":"DRAGON"})
p
```

```python
#5.standardize text
p["Name"]=p["Name"].str.lower()
p
```

```python
#6.fix data types
p["Legendary"]=p["Legendary"].astype(bool)
p
```

```python
#7. drop duplicates
p=p.drop_duplicates() #there were no duplicates
p
```
